```python
import numpy as np
import scipy
import math
import matplotlib.pyplot as plt
from bayes_opt import BayesianOptimization

# a = 2
# b = 3
# g = 5



def black_box_function(x, y, a=1.063, b=2.675, g=5):
    # return ((y) ** 2) * np.sin(a * x + b * y) + random.randint(1, 5)
    return ((y) ** 2) * np.sin(a*x + y*b)
    # return a*x**3*y**7 + b*x**2*y**8 - 4*x*y**9 + a*x**10 + b*y**10 + a*x**9*y + 8*x**2*y**2 - b*x**5*y**5 + 10

# Bounded region of parameter space
pbounds = {'x': (-7, 7), 'y': (-7, 7), 'a' : (0,6), 'b': (0,6), 'g': (0,6)}
optimizer = BayesianOptimization(
    f=black_box_function,
    pbounds=pbounds,
    random_state=0,
    verbose=2,
)

optimizer.maximize(
    init_points=5,
    n_iter=20
)
```

В данной реализации мы используем стандартную ядерную функцию, а именно Радиально-базисную:
k(x,x′)=exp((−∥x−x′∥)^2 / 2*l^2), где ∥x−x′∥ - норма Евклида между x и x′, а l - масштабный параметр.

```python
for ind, var in enumerate(optimizer.res):
    print({'Index: {}, Target: {}'.format(ind, var['target'])})
    print('{}'.format(var['params']), '\n')
```

Градиентный спуск - это итерационный метод оптимизации, который используется для нахождения минимума функции. Он основывается на понятии градиента, который является вектором частных производных функции по каждой переменной.

Идея градиентного спуска заключается в том, что мы начинаем с некоторого начального приближения точки минимума и затем итеративно изменяем это приближение, двигаясь в направлении, противоположном градиенту функции в данной точке. Таким образом, мы движемся вдоль крутых склонов функции в направлении минимума.

Алгоритм градиентного спуска можно описать следующим образом:

Задать начальную точку x_0.

Пока не выполнен критерий остановки, повторять шаги 3-6:

Вычислить градиент функции в текущей точке: g_k = grad f(x_k).

Выбрать направление движения: d_k = -g_k.

Выбрать длину шага (learning rate) alpha_k.

Обновить текущую точку: x_{k+1} = x_k + alpha_k*d_k.

Вернуть последнюю точку x_k+1 как приближение точки минимума функции.

Критерий остановки может быть разным, например, можно остановить итерацию, когда вектор градиента станет достаточно малым, или когда достигнут заданный предел числа итераций.

Одним из главных недостатков градиентного спуска является то, что он может застрять в локальных минимумах, не достигнув глобального минимума функции. Для преодоления этой проблемы могут использоваться различные модификации метода, такие как стохастический градиентный спуск, методы сопряженных градиентов и др.
